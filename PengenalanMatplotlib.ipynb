## Figure, axes, dan axis

Saat membuat sebuah graph/chart, pada matplotlib, terdapat paling tidak 3 objek berikut: figure, axes, dan axis. Berikut penjelasan singkatnya:

* Figure merupakan container dalam menampilkan grafik
* Axes merupakan area gambar. 
  - Plotting data dilakukan pada axes. 
  - Sebuah axes terdapat dalam figure. 
  - Sebuah figure dapat menampung beberapa axes. 
* Axis merupakan sumbu 
  - Graph 2D: 2 sumbu (x,y)
  - Graph 3D: 3 sumbu (x,y,z)

Terdapat beberapa cara dalam membuat objek figure dan axes. Dengan mengasumsikan *plt adalah import dari matplotlib.pyplot*, berikut beberapa kode yang mungkin dituliskan.

**Kode 1**: figure **tanpa axes**

```
fig = plt.figure()
```

**Kode 2**: sebuah figure **dengan sebuah axes**

```
fig, ax = plt.subplots()
```

**Kode 3**: sebuah figure **dengan 6 objek axes**

```
fig, axs = plt.subplots(2,3)
```
Kode di atas, akan membentuk 2x3 grid axes (2 baris 3 kolom). Parameter pertama menyatakan jumlah baris, dan parameter kedua menyatakan jumlah kolom.

## Menampilkan graph (pyplot style) 
Untuk menampilkan data (x,y) pada chart 2D sederhana dengan cara OO, dapat dilakukan dengan subplots() dan plot() dari matplotlib.pyplot. Pyplot merupakan kumpulan function yang memungkinkan pembuatan figure, plot titik (x,y) dan lainnya. 





````python
import matplotlib.pyplot as plt

# data (x,y) yang diplot
x = [-2, 0, 2, 4, 6, 8]
y = [-2, 0, 4, 5, -8, 10]

plt.suptitle('Judul Graph')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.plot(x,y)
````

Contoh lainnya:

````python
import matplotlib.pyplot as plt
import numpy as np

# data (x,y) yang diplot
x = np.linspace(0, 2, 100)
y1 = [a**2 for a in x]

plt.plot(x, x, label='linear')
plt.plot(x, y1, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.title('Simple plot')
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
plt.show()
````

Keterangan:

* Pada contoh di atas, "plt.plot" memerlukan 2 parameter mandatory (x,y), dan dapat ditambah parameter lain (label)
* "plt.suptitle" memberikan judul pada figure (main/utama), sedangkan "plt.title" memberikan judul pada tiap axes
* "plt.legend()" berfungsi untuk menampilkan legend
* "plt.xlabel" memberikan keterangan pada sumbu x
* "plt.ylabel" memberikan keterangan pada sumbu y
* Pembuatan "y1" menggunakan list comprehension bersifat optional dilakukan. Jika ingin dikodekan "lebih rapi", dapat dikodekan seperti pembuatan line plot cubic. 
* "plt.show()" menghilangkan teks yang muncul akibat command terakhir untuk membuat graph. Contoh teks yang dihilangkan pada graph pertama di atas adalah keterangan "[<matplotlib.lines.Line2D at 0x7f11e1b12fd0>]" 


## Menampilkan graph (OO style)

Cara lain untuk melakukan plotting, yaitu dengan menggunakan OO style. Pada style ini, terdapat kode yang ditulis untuk membentuk objek figure & axes terlebih dahulu. Setelah itu, dilakukan plotting titik (x,y) dengan memanggil method **plot dari axes** (bukan dari matplotlib.pyplot)

````python
import matplotlib.pyplot as plt

# data (x,y) yang diplot
x = [-2, 0, 2, 4, 6, 8]
y = [-2, 0, 4, 5, -8, 10]

fig, ax = plt.subplots()
ax.set_title("Judul Graph")
ax.set_xlabel("sumbu x")
ax.set_ylabel("sumbu y")
ax.plot(x,y)
````

Contoh lainnya:

````python
import matplotlib.pyplot as plt
import numpy as np

# data (x,y) yang diplot
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()

ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_title("Simple plot")
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.legend()
plt.show()
````
