# Menampilkan Gambar

Ada banyak cara menampilkan gambar/citra atau image dengan menggunakan bahasa Python, salah satunya dengan memanfaatkan Matplotlib. 

## Kasus
Tampilkan gambar dari path gdrive (asumsi gambar sudah ada). Path yang digunakan pada kasus ini adalah: "/content/Doraemon_character.png".

## Penyelesaian
Menggunakan 2 hal berikut: 
* imread dari matplot.image, dan 
* imshow dari matplotlib.pyplot. 

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('/content/Doraemon_character.png')
imgplot = plt.imshow(img)
plt.show()
```

Berikut beberapa kode lain yang dapat ditambahkan agar menampilkan selayaknya sebuah gambar:
* `plt.title('x')` menambah judul "x" gambar
* `set_frame_on(False)`  menghapus frame pada gambar
* `ax.get_xaxis().set_visible(False)`  menghapus axis text atau tickmark pada sumbu x
* `ax.get_yaxis().set_visible(False)`  menghapus axis text atau tickmark pada sumbu y


```python
img = mpimg.imread('/content/Doraemon_character.png')
imgplot = plt.imshow(img)
plt.title("Doraemon")

ax = plt.gca()
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.show()
```

## Menampilkan Gambar Side-by-side

Seperti halnya chart, citra juga dapat ditampilkan secara bersisian menggunakan subplot. 
Berikut adalah contoh menampilkan gambar secara side-by-side secara bersisian. 

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# ambil path
img_doraemon = mpimg.imread('/content/Doraemon_character.png')
img_dorami = mpimg.imread('/content/dorami.jpg')

fig, (ax1, ax2) = plt.subplots(1, 2)

# gambar pertama - doraemon
# ax1 = plt.subplot(1,2,1)
ax1.set_title('Doraemon')
ax1.imshow(img_doraemon)

# setting gambar pertama
ax1.set_frame_on(False)
ax1.get_xaxis().set_visible(False)
ax1.get_yaxis().set_visible(False)

# gambar kedua - dorami
ax2 = plt.subplot(1,2,2)
ax2.set_title('Dorami')
ax2.imshow(img_dorami)

# setting gambar kedua
ax2.set_frame_on(False)
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)

```
