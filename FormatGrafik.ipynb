# Pengaturan style/format grafik

Sebuah grafik yang dibentuk dengan pyplot dapat diatur bentuknya, seperti skala x. y maksimum, bentuk titik, warna garis, dan lainnya. 

## Menentukan "panjang" sumbu x/y
Untuk menentukan "jarak" garis sumbu x & y, dapat digunakan method `plt.axis([xmin, xmax, ymin, ymax])`.
* xmin → nilai minimum garis/sumbu x
* xmax → nilai maksimum garis x
* ymin → nilai minimum garis/sumbu y
* ymax → nilai maksimum garis y

## Menentukan format tampilan grafik
Format grafik ini ditentukan dari **parameter ketiga method plot**. Jadi bentuk umum pemanggilannya: `plt.plot(x,y,format)`. 
* x → nilai koordinat sumbu x
* y → nilai koordinat sumbu y
* format menyatakan bentuk tampilan garis/titik. 

Format dituliskan dalam bentuk string yang dinamakan fmt, terdiri dari `marker|linestyle|color`. Selain itu, format bisa juga dituliskan dalam bentuk parameter terpisah.
* marker → titik penanda koordinat (x,y)
* linestyle → bentuk garis
* color → warna dari garis

Jika format tidak ditentukan, nilai default format adalah '-b':
* \- → garis solid
* b → warna blue (biru)
* tidak ada marker digunakan

Contoh grafik dengan pengaturan penskalaan (plot.axis) tanpa pengaturan format dapat dilihat pada contoh di bawah:

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 3, 9]

plt.plot(x, y)

# atur skala x (0-6) & y (0-15)
plt.axis([0, 6, 0, 15])
plt.show
```

Penulisan string fmt, tidak harus mengikuti urutan `marker|linestyle|color`. Urutannya dapat dibalik, contohnya parameter ketiga dituliskan "--ro", "r--o" atau "or--" akan menghasilkan format yang sama: garis merah (r) dengan titik koordinat berbentuk lingkaran (o), dan garis putus-putus (--).

Pengaturan ini dapat dilihat pada dokumentasi [marker matplotlib](https://matplotlib.org/stable/api/markers_api.html) atau tutorial matplotlib (seperti [w3school](https://www.w3schools.com/python/matplotlib_markers.asp)).

```python
plt.subplot(1,3,1)
plt.title("format: --ro")
plt.plot(x, y, '--ro')

plt.subplot(1,3,2)
plt.title("format: o--r")
plt.plot(x, y, 'o--r')

plt.subplot(1,3,3)
plt.title("format: or--")
plt.plot(x, y, 'or--')

plt.plot()
```

Pengaturan grafik ini juga dapat diatur satu per satu. Contohnya dapat dilihat pada kode berikut:


```python
plt.plot(x, y, marker='D', color='g', linestyle='-.')
plt.plot()
```

## Format ukuran 

Hal di atas merupakan format basic yang dapat ditentukan untuk mengubah style tampilan sebuah grafik. Selain itu, ada juga hal lain yang dapat diatur pada pemanggilan `plot()`, seperti:
* Ukuran marker → parameter `markersize` atau `ms`. 
* Wana tepi marker → parameter `markeredgecolor` atau `mec`. 
* Warna isian marker → parameter `markerfacecolor` atau `mfc`. 

Contoh, kode di bawah ini menghasilkan marker berukuran 20 (ms = 20), dengan warna tepi merah (mec = 'r'), dan warna isian hijau (mfc = 'g').



```python
plt.plot(x, y, 'D:', ms = 20, mfc = 'g', mec = 'r')
plt.plot()
```

Untuk warna yang digunakan, dapat mengikuti referensi nilai heksadesimal, atau HTML color names (seperti hotpink, maroon, dll).
