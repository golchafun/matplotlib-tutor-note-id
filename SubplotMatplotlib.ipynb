# Menggunakan Subplot

Subplot digunakan untuk menampilkan beberapa grafik secara bersisian (side-by-side). Inti dari subplot adalah menggunakan **banyak figure untuk menampung beberapa axes**. Seperti yang kita tahu, "hirarki" grafik pada matplotlib adalah:
> **figure → axes → plotting data**.

Hal yang dilakukan pertama kali adalah menentukan berapa chart/plot yang ditampilkan dalam berapa baris & kolom data. Kode untuk menentukan hal ini adalah:
`plot.subplot(n, m, x)`
* n menyatakan jumlah baris
* m menyatakan jumlah kolom
* x menyatakan urutan grafik

Urutan **dimulai dari 1**, dan dihitung dari **kanan ke kiri - atas ke bawah**. 

Kode tersebut dituliskan sebelum melakukan plotting chart.

## Contoh kasus

Dibentuk 2 buah grafik berdampingan (kiri-kanan), satu grafik menampilkan hubungan x-x linear, dan grafik lainnya menampilkan hubungan kuadratik (x-x²). Angka x bernilai 0-20. 

## Penyelesaian (pyplot style)
Dibentuk 2 grafik yang ditampilkan **1 baris - 2 kolom**. Sehingga, kode subplot yang digunakan adalah `subplot(1,2,1)` untuk grafik pertama, dan `subplot(1,2,2)` untuk grafik kedua. 

Contoh kode adalah sebagai berikut:

```python
import matplotlib.pyplot as plt

# data (x,y) yang diplot
x = list(range(0,20))
y = [a**2 for a in x]

plt.suptitle('Linear & Kuadratik')

plt.subplot(1,2,1)
plt.title('Graph Linear')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.plot(x,x,label='linear')

plt.subplot(1,2,2)
plt.title('Graph Kuadratik')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.plot(x,y,label='kuadratik')

plt.show()
```

Grafik di atas memerlukan penyesuaian tampilan. Hal ini dikarenakan ada keterangan sumbu y dari grafik 2 yang overlap terhadap grafik 1. Untuk itu, diperlukan pengubahan tampilan grafik di atas, dengan pengaturan kode berikut:
* `plt.subplots_adjust(wspace=0.5, top=0.8)` → mengatur jarak antar chart & title dengan chart
* `plt.rcParams['axes.spines.top'] = False` → menghilangkah garis di atas pada frame chart
* `plt.rcParams['axes.spines.right'] = False` → menghilangkah garis di kanan pada frame chart

```python
plt.suptitle('Linear & Kuadratik')

# konfigurasi tampilan
plt.subplots_adjust(wspace=0.5, top=0.8)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

# chart 1
plt.subplot(1,2,1)
plt.title('Graph Linear')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.plot(x,x,label='linear')

# chart 2
plt.subplot(1,2,2)
plt.title('Graph Kuadratik')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.plot(x,y,label='kuadratik')

plt.show()
```

## Penyelesaian (OO style)

Selain dengan kode di atas, penggunaan subplot juga dapat dilakukan dengan menggunakan kode style object oriented (OO-style). 

```python
# atur grafik 1 baris 2 kolom
fig, ax = plt.subplots(1, 2)
fig.suptitle('Linear & Kuadratik') 

# chart 1
ax[0].plot(x,x)
ax[0].set_xlabel('sumbu x')
ax[0].set_ylabel('sumbu y')
ax[0].set_title('Graph Linear')

# chart 2
ax[1].plot(x,y)
ax[1].set_xlabel('sumbu x')
ax[1].set_ylabel('sumbu y')
ax[1].set_title('Graph Kuadratik')

# atur jarak (parameter boleh kosong)
fig.tight_layout(pad=5.0)

plt.show()
```

## Keterangan tambahan

### subplot.adjust
Beberapa nilai parameter dari subplot.adjust yang dapat diatur untuk mengubah jarak antar chart adalah:
* left
* right
* top
* bottom
* wspace
* hspace

Nilai ini juga dapat dilihat dengan kode `plt.subplot_tool()`. 

### rcParams
Nilai rcParams merupakan properti spines dari sebuah axes. Ada 4 nilai di dalamnya:
* left
* right
* bottom
* top

Contoh pengaksesannya dapat dilihat pada contoh kode di bawah.
